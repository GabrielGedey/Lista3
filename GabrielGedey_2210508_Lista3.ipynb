```python
import pandas as pd
from google.colab import drive


drive.mount('/drive')
passageiros = pd.read_csv("/drive/MyDrive/class_titanic.csv")

passageiros.info()
display(passageiros)
```

```python
passageiros.dropna(inplace = True,subset = ['Age'])
display(passageiros)
```

```python
media = passageiros["Age"].mean()
display(passageiros)
print("%.1f" %media)
```

```python
media
#Dataframe novo sem idades com variação maior que 15 em relação a media
passageiros2 = passageiros.loc[(passageiros["Age"] <= (media + 15)) & (passageiros["Age"] >= (media - 15))]
display(passageiros2)
```

```python
passageiros_new = passageiros2.drop(["PassengerId","Sex","SibSp","Parch","Ticket","Cabin","Embarked"],axis = 1)
display(passageiros_new)
```

```python
import random as r
passageiros_new["Pagamento_classe"] = ((passageiros_new["Fare"] - passageiros_new["Pclass"]) * r.uniform(0,1))
display(passageiros_new)
```

```python
import numpy as np
passageiros_new = passageiros_new.reindex(np.random.permutation(passageiros_new.index))
display(passageiros_new)
```

```python
#Ordena o Dataframe por sobrenome em ordem alfabetica
passageiros_new = passageiros_new.sort_values(by='Name', ascending=True)
display(passageiros_new)
```

```python
grupo = passageiros_new.groupby('Name')[['Pagamento_classe']].mean()
display(grupo)
```

```python
passageiros_new["Survived"] = passageiros_new["Survived"].map({0: "Não",1: "Sim"})
display(passageiros_new)
```

```python
passageiros_new.to_csv('/drive/My Drive/passageiros_new.csv')
```
